# Redução de dimensionalidade

Programa de redução de dimensionalidade. Imagem em RGB, levada a tons de cinza depois a preto e branco.

```python
import numpy as np
import requests
from io import BytesIO
import matplotlib.pyplot as plt
from PIL import Image

url = 'https://img.pokemondb.net/artwork/large/ivysaur.jpg'

response = requests.get(url)
img_bytes = BytesIO(response.content)

imagem = Image.open(img_bytes).convert('RGB')


pixels = np.array(imagem)


r = pixels[:, :, 0]
g = pixels[:, :, 1]
b = pixels[:, :, 2]

gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
gray = gray.astype(np.uint8)

# 👁️ Exibir imagem original e em tons de cinza
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(pixels)
axs[0].set_title('Imagem Original')
axs[0].axis('off')

axs[1].imshow(gray, cmap='gray')
axs[1].set_title('Imagem em Tons de Cinza')
axs[1].axis('off')

plt.show()
```

**Após a transição da imagem em RGB para tons de cinza, é feita a transição de tons de cinza, para preto e branco.**

```python
limiar = 160
bw = np.where(gray > limiar, 255, 0).astype(np.uint8)


fig, axs = plt.subplots(1, 3, figsize=(12, 6))

axs[0].imshow(pixels)
axs[0].set_title('Imagem Original')
axs[0].axis('off')

axs[1].imshow(gray, cmap='gray')
axs[1].set_title('Imagem em Tons de Cinza')
axs[1].axis('off')

axs[2].imshow(bw, cmap='gray')
axs[2].set_title('Imagem em Preto e Branco')
axs[2].axis('off')


plt.show()

```
